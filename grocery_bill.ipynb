```python
##### print grocery bill
#use multi fdimesnsional lsit
items_tot=int(input("Enter total number of items purchased"))
total= 0
item=[]
for i in range(0,items_tot):
    name=input("Enter item name")
    qnt=int(input("enter quantity purchased"))
    price=int(input("enter price of item purchased"))
    total=total+(qnt*price)
    item.append([name,qnt,price])
                     
print("*"*16,"BILL","*"*16)
print("Item Name \t Item Quantity \t Item Price")
for k in item:
    print(f"{k[0]}\t\t{k[1]}\t\t{k[2]}")# f here stands for formatted string that allows ypu to add variable,list ,tuple withib curly braces
print("*"*40)
print("Total Amount to be paid is:",total)
print("*"*40)
```
